```python
# Defining the function to display central tendencies
def central_tendencies(data, column_name):
    # Calculate central tendencies
    mean_value = data[column_name].mean()
    mode_value = data[column_name].mode()[0]  # mode() returns a series, so use [0] to get the first mode
    median_value = data[column_name].median()
    
    # Print results in the specified format
    print(f"data['{column_name}'] - measures of central tendencies")
    print('='*50)
    print(f'Mean of the column is {mean_value}')
    print('='*50)
    print(f'Mode of the column is {mode_value}')
    print('='*50)
    print(f'Median of the column is {median_value}')

# Testing the function with 'claim size' column
central_tendencies(data, 'claim size')
```

#### Insights
- **Mean** of the column is 2560.5: The average claim size is 2560.5
- **Mode** of the column is 2300: The most frequently occurring claim size is 2300 units. This indicates that many policies have a claim size around this value.
- **Median** of the column is 2600.0: The middle value of the claim sizes is 2600. This means that half the claim sizes are below 2600, and half are above.

The central tendency results suggest that the data is relatively symmetrically distributed around the 2300â€“2600 range, with most claim sizes being close to the mean, median, and mode. This suggests there isn't much skew or extreme outliers in the central portion of the distribution.

```python
# Defining the function to display measures of spread (min, max, range, standard deviation, variance) 
def spread_measures(data, column_name):
    # Calculate measures of spread
    mean_value = data[column_name].mean()
    mode_value = data[column_name].mode()[0]  # mode() returns a series, so use [0] to get the first mode
    median_value = data[column_name].median()
    
    min_value = data[column_name].min()
    max_value = data[column_name].max()
    range_value = data[column_name].max() - data[column_name].min()
    variance_value = round(data[column_name].var(), 2)
    std_value = round(data[column_name].std(), 2)
    
    # Print results in the specified format
    print(f"data['{column_name}'] - measures of spread")
    print('='*50)
    print(f'Min of the column is {min_value}')
    print('='*50)
    print(f'Max of the column is {max_value}')
    print('='*50)
    print(f'Range of the column is {range_value}')
    print('='*50)
    print(f'Variance of the column is {variance_value}')
    print('='*50)
    print(f'Standard deviation of the column is {std_value}')

# Testing the function with 'claim size' column
spread_measures(data, 'claim size')
```

#### Insights
- **Min** of the column is 200: The smallest claim size in the dataset is 200 units.
- **Max** of the column is 5000: The largest claim size is 5000 units.
- **Range** of the column is 4800: The difference between the smallest and largest claim sizes is 4800 units, showing that there is a wide spread in claim sizes.
- **Variance** of the column is 1921411.16: Variance indicates how much the claim sizes differ from the mean. A higher variance indicates that the data points are more spread out.
- **Standard deviation** of the column is 1386.15: Standard deviation provides a measure of the average distance of each claim size from the mean. With a standard deviation of 1386.15, it suggests that most claim sizes deviate significantly from the mean, indicating substantial variability in the data.

### 1.1.2 Calculating Pearson correlation coefficient

The Pearson correlation coefficient (also known as Pearson's r or simply r) is a measure of the linear relationship between two variables. It quantifies how closely the two variables are related in terms of how one variable changes in response to changes in the other. The Pearson correlation coefficient has the following properties:

**Range:**
- The coefficient ranges from -1 to 1.
- **+1** means a perfect positive linear correlation (as one variable increases, the other increases in a perfectly proportional way). (×›×›×œ ×©××©×ª× ×” ××—×“ ×’×“×œ, ×”×©× ×™ ×’×“×œ ×‘××•×¤×Ÿ ×¤×¨×•×¤×•×¨×¦×™×•× ×œ×™ ×œ×—×œ×•×˜×™×Ÿ)
- **-1** means a perfect negative linear correlation (as one variable increases, the other decreases in a perfectly proportional way). (×›×›×œ ×©××©×ª× ×” ××—×“ ×’×“×œ, ×”×©× ×™ ×™×•×¨×“ ×‘×¦×•×¨×” ×¤×¨×•×¤×•×¨×¦×™×•× ×œ×™×ª ×œ×—×œ×•×˜×™×Ÿ)
- **0** means no linear correlation (there is no discernible linear relationship between the variables).

**Magnitude:**
- The **closer the value of r is to 1 or -1**, the stronger the relationship.
- If the coefficient is close to **0**, it means there is little or no linear relationship between the variables.

**Interpretation:**
- **r = 1**: Perfect positive correlation. (××ª×× ×—×™×•×‘×™ ××•×©×œ×)
- **r = -1**: Perfect negative correlation. (××ª×× ×©×œ×™×œ×™ ××•×©×œ×)
- **r = 0**: No correlation. (××™×Ÿ ×§×•×¨×œ×¦×™×”)
- **0.1 < |r| < 0.3**: Weak correlation. (××ª×× ×—×œ×©)
- **0.3 < |r| < 0.5**: Moderate correlation. (××ª×× ××ª×•×Ÿ)
- **0.5 < |r| < 1**: Strong correlation. (××ª×× ×—×–×§)

**Usage:**
Pearson's r is commonly used in statistics to assess the strength and direction of the linear relationship between two continuous variables.

```python
# Calculating Pearson correlation coefficient between 'claim size' and 'vetek'
correlation = data['claim size'].corr(data['vetek'], method='pearson')
correlation
```

- Market share of car insurance: Î© = {C, M, P, H}
     C= 40%, M=35%, P=15%, H=10%
- The market share is the probability to be insured in each company in the market
- The market share of car insurance is comprised of 4 mutually exclusive events (because no one will pay twice for car insurance)
- The probability for a total loss:
    TotalÂ Lossâˆ£C= 2%, TotalÂ Lossâˆ£M=1%, TotalÂ Lossâˆ£P=4%, TotalÂ Lossâˆ£H=5% - the conditional probabilities for total loss, given a selected insurance company

### Q1 Solution

To find the probability of a total loss in car insurance using the law of total probability, we need to use the formula:

    P(TotalÂ Loss)= P(C)â‹…P(TotalÂ Lossâˆ£C) + P(M)â‹…P(TotalÂ Lossâˆ£M) + P(P)â‹…P(TotalÂ Lossâˆ£P) + P(H)â‹…P(TotalÂ Lossâˆ£H)

Where:

- P(TotalÂ Lossâˆ£C), P(TotalÂ Lossâˆ£M, P(TotalÂ Lossâˆ£P), P(TotalÂ Lossâˆ£H): conditional probabilities of total loss given the company
     

```python
# Given data
market_shares = {
    'C': 0.40,
    'M': 0.35,
    'P': 0.15,
    'H': 0.10
}

conditional_probabilities = {
    'C': 0.02,
    'M': 0.01,
    'P': 0.04,
    'H': 0.05
}

# Law of total probability calculation
total_loss_probability = sum(market_shares[company] * conditional_probabilities[company] for company in market_shares)
total_loss_probability
```

### Cumulative Distribution Function

The cumulative distribution function helps us answer the question:
- what is the probability that the random variable ğ‘‹ will be **smaller or equal** to a certain value?
$$
    ğ‘ƒ_{ğ‘‹}(ğ‘¥) = ğ‘ƒ(ğ‘‹ â‰¤ ğ‘¥)
$$

The CDF accumulates probabilities along the values of the random variable
- It starts at 0, before any values have been reached
- It adds the probability of each value, as more values of the random variable are met
- It ends at one, after all probabilities have been accumulated

Question:
- What is the probability of X being smaller or equal to 9?
<div align="left">

| Values of X | 0 | 1 | 2 | 3 | 4 |
|:---|:---|:---|:---|:---|:---|
| Probabilities | 0.1 | 0.2 | 0.3 | 0.3 | 0.1 |

</div>
For the given probabilities:
    
P(X=0)=0.1
    
P(Xâ‰¤1)=0.1+0.2=0.3
    
P(Xâ‰¤2)=0.1+0.2+0.3=0.6
    
P(Xâ‰¤3)=0.1+0.2+0.3+0.3=0.9
    
P(Xâ‰¤4)=0.1+0.2+0.3+0.3+0.1=1.0

Since we only have probabilities up to X=4, 
the cumulative probability remains at 1 for all values greater than 4.

```python
import numpy as np
import matplotlib.pyplot as plt

# Defining Values and Probabilities
values = [0, 1, 2, 3, 4]
probabilities = [0.1, 0.2, 0.3, 0.3, 0.1]

# Calculate cumulative probabilities (CDF)
cumulative_probabilities = np.cumsum(probabilities)

# Extend values and cumulative probabilities to answer the question about X <= 9
extended_values = list(range(0, 10))  # From 0 to 9
# Extend the cumulative probabilities by appending 1.0 (the last value of the CDF) for each value from 5 to 9.
extended_cumulative_probabilities = [cumulative_probabilities[-1]] * (len(extended_values) - len(values))
extended_cumulative_probabilities = list(cumulative_probabilities) + extended_cumulative_probabilities

# Plot the CDF
plt.step(extended_values, extended_cumulative_probabilities, where='post', color='red', marker='o')

# Adding labels and title
plt.xlabel('Values of X')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF)')

# Answering the question - Probability of X <= 9
prob_X_leq_9 = extended_cumulative_probabilities[-1]
print(f"Probability of X being smaller or equal to 9 is: {prob_X_leq_9}")

# Show the plot
plt.grid(True)
plt.show()
```

### Binomial Random Variable

**The Binomial random variable** counts the number of successes in a series of ğ‘› independent Bernoulli experiments with a constant probability of success ğ‘ in each experiment 

- **Binomial random variable**: ğ‘¿~ğ‘©ğ’Šğ’(ğ’, ğ’‘) = number of successes
- **Possible values**: 0,1, â€¦ , n
- **Parameters of the distribution**: ğ‘› â€“ number of experiments, ğ‘ â€“ the probability of success
- **Probability distribution function**:
$$
P(X = x) = \binom{n}{x} \cdot p^x \cdot (1 - p)^{n - x}
$$
Where:
- \( n \) is the number of trials,
- \( p \) is the probability of success on a single trial,
- \( x \) is the number of successes.
- **Mean:** nÃ—p



Suppose two binomial random variables, with n=15 and the following probabilities:
- ğ‘ = 0.9
- ğ‘ = 0.2

What are your expectations about the PDF and the CDF of each variable?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 15 # Number of trials
p1 = 0.9  # Probability for first binomial variable
p2 = 0.2  # Probability for second binomial variable
x = np.arange(0, n+1)  # Possible outcomes of X (from 0 to 15)

# Calculate PDF for both binomial distributions
# This function calculates the probability mass function (PMF), 
# which gives the probability for each possible outcome x
# for a binomial distribution with parameters n and p
pdf_p1 = binom.pmf(x, n, p1)
pdf_p2 = binom.pmf(x, n, p2)

# Calculate CDF for both binomial distributions
# This function calculates the cumulative distribution function (CDF), 
# which gives the probability that X will take a value less than or equal to x
cdf_p1 = binom.cdf(x, n, p1)
cdf_p2 = binom.cdf(x, n, p2)

# Create the figure and subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot PDF
axs[0].bar(x - 0.2, pdf_p1, width=0.4, label=f'PDF (p={p1})', color='blue', align='center')
axs[0].bar(x + 0.2, pdf_p2, width=0.4, label=f'PDF (p={p2})', color='orange', align='center')
axs[0].set_title('Probability Distribution Function (PDF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('P(x)')
axs[0].legend()

# Plot CDF
# We use step() for CDFs since the CDF is a step function, 
# where each step represents the cumulative probability up to that point.
axs[1].step(x, cdf_p1, where='post', label=f'CDF (p={p1})', color='blue')
axs[1].step(x, cdf_p2, where='post', label=f'CDF (p={p2})', color='orange')
axs[1].set_title('Cumulative Distribution Function (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].legend()

# Display the plots
plt.tight_layout()
plt.show()
```

#### Interpretation:
1. **PDF (Probability Distribution Function):**
     - For p=0.9, the distribution is **right-skewed**, meaning that most of the probability mass is concentrated near higher values of 
    - For p=0.2, the distribution is **left-skewed**, meaning that most of the probability mass is concentrated near lower values of 

2. **CDF (Cumulative Distribution Function):**

**SOLUTION A**

    $$
        E(X)=nâ‹…p

```python
import numpy as np
from scipy.stats import binom

# Product A parameters
n_A = 200  # Number of potential clients
p_A = 0.3  # Conversion probability
price_A = 2500  # Price per client
expenses_A = 800  # Expenses per offer

# Product B parameters
n_B = 100  # Number of potential clients
p_B = 0.4  # Conversion probability
price_B = 2800  # Price per client
expenses_B = 1100  # Expenses per offer

# Function to calculate profit for each possible number of conversions
def calculate_profit(n, p, price, expenses):
    # Array of possible conversions (0 to n)
    conversions = np.arange(0, n + 1)
    
    # Calculate the profit for each number of conversions
    profit_per_conversion = (price - expenses)
    profit = conversions * profit_per_conversion
    
    # Calculate binomial probabilities for each conversion count
    probabilities = binom.pmf(conversions, n, p)
    
    # Calculate expected profit
    expected_profit = np.sum(probabilities * profit)
    
    return expected_profit, probabilities, profit

# Calculate for Product A
expected_profit_A, probabilities_A, profit_A = calculate_profit(n_A, p_A, price_A, expenses_A)

# Calculate for Product B
expected_profit_B, probabilities_B, profit_B = calculate_profit(n_B, p_B, price_B, expenses_B)

# Output results
print(f"Expected profit for Product A: ${expected_profit_A:,.2f}")
print(f"Expected profit for Product B: ${expected_profit_B:,.2f}")
```

```python
from math import comb

# Given values
N = 24  # Total number of companies
K = 12  # Companies with rising stock value
n = 5   # Sample size

# (a) P(X = 2)
k = 2
p_x_2 = (comb(K, k) * comb(N - K, n - k)) / comb(N, n)

# (b) P(X <= 2)
p_x_0 = (comb(K, 0) * comb(N - K, n - 0)) / comb(N, n)
p_x_1 = (comb(K, 1) * comb(N - K, n - 1)) / comb(N, n)
p_x_leq_2 = p_x_0 + p_x_1 + p_x_2

# (c) Expected value (mean)
E_X = n * (K / N)

# (d) Variance
Var_X = n * (K / N) * ((N - K) / N) * ((N - n) / (N - 1))

# Output the results
print(f"P(X = 2): {p_x_2:.4f}")
print(f"P(X <= 2): {p_x_leq_2:.4f}")
print(f"Expected value E(X): {E_X:.2f}")
print(f"Variance Var(X): {Var_X:.2f}")
```

```python
# Defining the probability and outcome functions for the roulette game
def expected_value_color(bet_amount):
    # Probability of winning on color
    p_win_color = 18 / 37
    # Expected value for betting on a color
    return (p_win_color * 2 * bet_amount)

def expected_value_number(bet_amount):
    # Probability of winning on a specific number
    p_win_number = 1 / 37
    # Expected value for betting on a specific number
    return (p_win_number * 36 * bet_amount)

# Example bet amount (can be changed)
bet_amount = 1

# Calculate the expected values
expected_value_specific_color = expected_value_color(bet_amount)
expected_value_specific_number = expected_value_number(bet_amount)

print(f"Expected Values of Specific Color: {expected_value_specific_color:.2f}")
print(f"Expected Values of Specific Number: {expected_value_specific_number:.2f}")
```

The roulette wheel has 37 pockets: 18 red, 18 black, and 1 green (0). This creates an inherent imbalance in the game because the odds are not in favor of the player in the long run. Regardless of whether you bet on a color or a number, the house retains a slight advantage due to the presence of the "0" pocket.
2. **Expected Value Formula:**<br>
Expected value is calculated as:
$$
    E=(ProbabilityÂ ofÂ WinÃ—Payout)+(ProbabilityÂ ofÂ LossÃ—PayoutÂ ofÂ Loss)
$$
3. **Bet on Color:**<br>
    - Probability of winning on color (black or red) is 18/37
    - If you win, you get paid 2x your bet (because the payout for color is 2:1).
    - Expected value formula for betting on color is:
    $$
       E_{color} = (\frac{18}{37} Ã— 2X) + (\frac{19}{37} Ã— 0) = \frac{36X}{37}
    $$
4. **Bet on Number:**<br>
    - Probability of winning on a specific number (like 17) is 1/37
    - If you win, you get paid 36x your bet (because the payout for a number is 36:1).
    - Expected value formula for betting on color is:
    $$
       E_{number} = (\frac{1}{37} Ã— 36X) + (\frac{36}{37} Ã— 0) = \frac{36X}{37}
    $$
5.**Why the Same?**<br>
    - Both bets yield an expected return of 36/37 of the original bet. In both cases, the payouts and the probabilities balance each other out in such a way that the expected value over time ends up being roughly 97.3% of the bet amount.<br> This reflects the **house edge** in roulette, where the presence of the "0" pocket ensures that, on average, the casino always wins a small portion of each bet.

## 6.1 <a id="continuous_random_variables">Continuous random variables</a> <font size='3'>[[Table of contentsğŸ”]](#index)]<font size>
    
A **discrete random variable** can get **specific distinct** values within a specified interval
    
A **continuous random variable** can have **any** value within a specified interval
    

#### Continuous Vs. Discrete
- The total probability is always 1
- A probability distribution function assigns probabilities to **discrete** values:
    - Each value ğ‘¥ of the discrete random variable ğ‘‹ is assigned a probability â€“ reducing the overall probability that can be assigned to other values
- Within a **continuous** interval there is an infinite amount of numbers!
    - We cannot assign a positive value of probability to each of the infinite numbers within the interval (no matter how small), and still expect that the overall probability will be 1
    - That is why the **probability of each specific value is 0** 

## 6.2 <a id="probability_density_function">Probability Density Function</a> <font size='3'>[[Table of contentsğŸ”]](#index)]<font size>
- So how do we assign probabilities?
    - A continuous random variable is **assigned probability by intervals** (that have length), not by specific values
For example, temperature is a continuous variable. We take a temperature measurement:
- What is the probability that the temperature will be exactly 25 degrees?
    - The probability of any specific value is 0
- The probability of the temperature being between 0-30 degrees (interval) is 0.9, the probability of the temperature being over 30 degrees is 0.1. 


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import poisson, norm

# Create a figure with 1 row and 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: Probability distribution for discrete variables (Poisson distribution)
x_poisson = np.arange(0, 11)
poisson_pmf = poisson.pmf(x_poisson, mu=3)  # Poisson distribution with lambda=3
ax1.stem(x_poisson, poisson_pmf, basefmt=" ", use_line_collection=True)
ax1.set_title(r'$P_X(x)$', fontsize=16)
ax1.set_xlabel('X', fontsize=14)
ax1.set_ylabel('Probability function', fontsize=14)
ax1.set_ylim(0, 0.5)

# Plot 2: Probability density function for continuous variables (Normal distribution)
x_norm = np.linspace(0, 10, 1000)
norm_pdf = norm.pdf(x_norm, loc=3, scale=1.5)  # Normal distribution with mean=3, std=1.5
ax2.plot(x_norm, norm_pdf, color='b')
ax2.fill_between(x_norm, norm_pdf, where=((x_norm >= 5) & (x_norm <= 6)), color='orange', alpha=0.5)
ax2.set_title(r'$f_X(x)$', fontsize=16)
ax2.set_xlabel('X', fontsize=14)
ax2.set_ylabel('Density function', fontsize=14)

# Set layout and display the plot
plt.tight_layout()
plt.show()
```

- The probability density function $$ ğ’‡_{ğ‘¿}(ğ’™) $$ for continuous variables is used to assign probabilities to intervals
-  $$ ğ’‡_{ğ‘¿}(ğ’™) $$ is analogous to the probability distribution function $$ P_{ğ‘¿}(ğ’™) $$ for discrete variables
- $$ ğ’‡_{ğ‘¿}(ğ’™) $$ assigns probability to an interval by calculating the **area** under the density curve

## 6.3 <a id="cumulative_distribution_function_&_quantiles_of_distribution">Cumulative distribution function & Quantiles of a distribution</a> <font size='3'>[[Table of contentsğŸ”]](#index)]<font size>

```python
# Define the density function parameters
x = np.linspace(0, 1, 100)
y = 2 * x

# Create the plot for the density function
plt.figure(figsize=(6, 6))
plt.plot(x, y, color='red', label=r'$f_X(x) = 2x$')
plt.fill_between(x, y, color='blue', alpha=0.3)

# Adding grid, labels and title
plt.grid(True)
plt.xlabel('x', fontsize=14)
plt.ylabel(r'$f_X(x)$', fontsize=14)
plt.title(r'Continuous Random Variable $f_X(x)$', fontsize=16)

# Limit the axis for clear view
plt.xlim(0, 1)
plt.ylim(0, 2)

# Show plot
plt.show()
```

Here is the recreated plot for the continuous random variable with the density function 
$$ ğ’‡_{ğ‘¿}(ğ’™) =2x\ for\ 0â‰¤xâ‰¤1 $$. The area under the curve represents the probability, and the shaded region illustrates the cumulative probability.

- Is it possible to have $$ ğ’‡_{ğ‘¿}(ğ’™) > 1 $$?
    - Yes. Density is not probability
- Calculate the area under the density plot
- The area is a triangle: $$ \frac{1 \cdot 2}{2} = 1 $$ 
- Calculate P(ğ‘‹ â‰¤ 0.5)
     - The area is a triangle: $$ \frac{0.5 \cdot 1}{2} = 0.25 $$

```python
# Create three plots demonstrating the cumulative distribution function (CDF) concept
x1 = np.linspace(0, 5, 500)
fx1 = np.exp(-0.5 * x1) * np.sin(x1) + 1  # Example density function

# Create subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# Plot 1: P(X â‰¤ a)
a = 2
ax1.plot(x1, fx1, color='blue')
ax1.fill_between(x1, fx1, where=(x1 <= a), color='blue', alpha=0.3)
ax1.set_title(r'$P(X \leq a) = F_X(a)$', fontsize=14)
ax1.set_xlabel('x')
ax1.set_ylabel(r'$f_X(x)$')

# Plot 2: P(b â‰¤ X â‰¤ c)
b, c = 1.5, 3
ax2.plot(x1, fx1, color='blue')
ax2.fill_between(x1, fx1, where=((x1 >= b) & (x1 <= c)), color='blue', alpha=0.3)
ax2.set_title(r'$P(b \leq X \leq c) = F_X(c) - F_X(b)$', fontsize=14)
ax2.set_xlabel('x')
ax2.set_ylabel(r'$f_X(x)$')

# Plot 3: P(X â‰¥ d)
d = 3
ax3.plot(x1, fx1, color='blue')
ax3.fill_between(x1, fx1, where=(x1 >= d), color='blue', alpha=0.3)
ax3.set_title(r'$P(X \geq d) = 1 - F_X(d)$', fontsize=14)
ax3.set_xlabel('x')
ax3.set_ylabel(r'$f_X(x)$')

# Set layout and display
plt.tight_layout()
plt.show()
```

## 6.4 <a id="quantiles_of_a_distribution">Quantiles of a Distribution</a> <font size='3'>[[Table of contentsğŸ”]](#index)]<font size>

```python
# Generating a bimodal distribution for the quantile plot example
np.random.seed(0)
data = np.concatenate([np.random.normal(0, 1, 500), np.random.normal(5, 1, 500)])

# Compute the quantiles
q25 = np.quantile(data, 0.25)
q50 = np.quantile(data, 0.5)
q75 = np.quantile(data, 0.75)

# Plot the distribution with quantiles
sns.kdeplot(data, color="blue")

# Add vertical lines for the quantiles
plt.axvline(q25, color='r', linestyle='--', label=r'$x_{0.25}$')
plt.axvline(q50, color='r', linestyle='--', label=r'$x_{0.50}$')
plt.axvline(q75, color='r', linestyle='--', label=r'$x_{0.75}$')

# Add labels and title
plt.title('Quantiles of a Distribution', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend()

# Display the plot
plt.show()
```

## 6.5 <a id="mean_and_variance">Mean and variance</a> <font size='3'>[[Table of contentsğŸ”]](#index)]<font size>

Mean and variance have similar definitions to the discrete case

**Mean:** 
$$
E(X) = \int_x x \cdot f_X(x) \, dx
$$
- Representing the center of the distribution
- Calculated as the weighted average of values (weighted by â€œprobabilitiesâ€)  

**Variance:** 
$$
V(X) = \int_x (x - E(X))^2 \cdot f_X(x) \, dx
$$
- Representing the spread of the distribution
- Calculated as the weighted average of squared distances of values from the mean (weighted by â€œprobabilitiesâ€)
- Since we will not be calculating integrals today, we will only use known means and variances of special continuous random variables

### The Uniform Distribution

```python
# Plot the uniform distribution example
a = 2
b = 6
height = 1 / (b - a)

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot([a, a, b, b], [0, height, height, 0], color="blue")  # Uniform distribution line
plt.fill_between([a, b], height, color='blue', alpha=0.2)      # Shaded area between a and b

# Add labels and title
plt.text(a, height/2, r'$\frac{1}{b-a}$', ha='right', fontsize=14)
plt.text(a, -0.05, 'a', ha='center', fontsize=12)
plt.text(b, -0.05, 'b', ha='center', fontsize=12)
plt.title('Uniform Distribution', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel(r'$f_X(x)$', fontsize=14)

# Set axis limits
plt.xlim(0, 8)
plt.ylim(0, 0.4)

# Display the plot
plt.grid(True)
plt.show()
```

The (continuous) Uniform random variable assigns an equal probability to intervals of equal length
within the interval (ğ‘,b)

Real life examples of the uniform distribution?
- Waiting in a stop light
- Waiting for a train to come

In addition, the uniform distribution plays an important role in simulation methods

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parameters for uniform distribution
a = 2
b = 8

# Create data for uniform distribution
x = np.linspace(a - 2, b + 2, 1000)
y_pdf = uniform.pdf(x, loc=a, scale=b - a)
y_cdf = uniform.cdf(x, loc=a, scale=b - a)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 8))

# Plot the PDF
ax1.plot(x, y_pdf, color='blue')
ax1.set_title('Density function $f(x)$', fontsize=14)
ax1.set_xlabel('x')
ax1.set_ylabel('$f_X(x)$')
ax1.grid(True)
ax1.set_ylim(0, 0.2)

# Plot the CDF
ax2.plot(x, y_cdf, color='blue')
ax2.set_title('Cumulative distribution function $F(x)$', fontsize=14)
ax2.set_xlabel('x')
ax2.set_ylabel('$F_X(x)$')
ax2.grid(True)
ax2.set_ylim(0, 1.1)

# Set layout and display
plt.tight_layout()
plt.show()
```

- **Range of values**
    - (a,b)
- **Parameters**
    - a min value
    - b max value
- **PDF of the uniform distribution:**
    $$
    f_X(x) = \frac{1}{b - a}, \quad a \leq x \leq b
    $$

- **CDF of the uniform distribution:**
    $$
    F_X(x) = \frac{x - a}{b - a}, \quad a \leq x \leq b
    $$

- **Mean of the uniform distribution:**
    $$
    E[X] = \frac{a + b}{2}
    $$
    
- **Shape**
    - symmetric

### Simulation

we want to generate random variables from a random variable with a known CDF $$ğ¹_{ğ‘‹}(x)$$
- Sample a number between 0-1 from a uniform distribution
- Pretend this number is a probability: $$ğ¹_{ğ‘‹}(x)$$ being smaller or equal to a value x
- By using the inverse connection between  $$ğ¹_{ğ‘‹}(x)$$ and quantiles â€“ find the suitable value from the desired distribution

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm

# Create a figure with 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot 1: Uniform distribution U(0,1)
x_uniform = np.linspace(0, 1, 1000)
y_uniform = uniform.pdf(x_uniform, loc=0, scale=1)

ax1.plot(x_uniform, y_uniform, color='blue')
ax1.set_title('U(0,1) density function $f(x)$', fontsize=14)
ax1.set_xlabel('x')
ax1.set_ylabel('$f_X(x)$')
ax1.grid(True)

# Highlight a point on the uniform distribution (for example, 0.25)
ax1.scatter(0.25, uniform.pdf(0.25), color='green', s=100, label="Sample point")
ax1.legend()

# Plot 2: Inverse transformation on CDF
x_values = np.linspace(-40, 60, 1000)
cdf_values = norm.cdf(x_values, loc=10, scale=10)

ax2.plot(x_values, cdf_values, color='blue')
ax2.set_title('F(x) - Cumulative distribution function', fontsize=14)
ax2.set_xlabel('x')
ax2.set_ylabel('F_X(x)')
ax2.grid(True)

# Highlight the corresponding CDF value and its inverse
ax2.hlines(0.25, -40, norm.ppf(0.25, loc=10, scale=10), color='green')
ax2.vlines(norm.ppf(0.25, loc=10, scale=10), 0, 0.25, color='green')
ax2.scatter(norm.ppf(0.25, loc=10, scale=10), 0.25, color='green', s=100, label="Mapped point")
ax2.legend()

# Display the plots
plt.tight_layout()
plt.show()
```

### The Exponential Distribution
- The Exponential random variable is a continuous random variable that is used to model the time between events, where the average rate of events per unit of time is ğœ† (Lambda)

- The exponential variable has an interesting and unique property known as the memoryless property
 $$ ğ‘ƒ(ğ‘‹ > ğ‘  + ğ‘¡|ğ‘‹ > ğ‘ ) = ğ‘ƒ(ğ‘‹ > t) $$
 
For example, if waiting time is exponential:

### Poisson distribution vs exponential distribution

1. #### Poisson Distribution:
    **What it describes:** The Poisson distribution models the number of events (like the number of potholes) occurring within a fixed interval of time or space (like 1 km of road or 1 minute of time).<br>
    **Example:** If potholes appear randomly on a 1 km road at an average rate of 1.6 potholes per kilometer, the Poisson distribution can tell you the probability of finding exactly 2 potholes in 1 km.<br>
    **Key parameter**: Î» (the average number of events per interval).<br>
    **Probability mass function (PMF):**
    $$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$
    Where:
    - X - number of events (e.g., number of potholes).
    - k - specific number of events (e.g., 2 potholes).
    - Î» - average number of events in the interval.
    
2. #### Exponential Distribution:
    **What it describes:** The exponential distribution models the time or distance between events in a Poisson process.<br>
    **Example:** If potholes appear randomly on a road, the exponential distribution can tell you the probability that the distance between successive potholes is more than 1 km.<br>
    **Key parameter:** Î» (the rate at which events occur, which is the inverse of the average time/distance between events).<br>
    **Probability density function (PDF):**
    $$
        f(x)=Î»e^{âˆ’Î»x} ,xâ‰¥0
    $$
    Where:
    - X - time/distance between two events (e.g., distance between potholes).
    - Î» - rate parameter (the average number of events per unit time/distance).  
    
#### Key Differences:
1. What they measure:
    - **Poisson distribution:** Measures the number of events in a fixed interval (like 2 potholes in 1 km).
    - **Exponential distribution:** Measures the time or distance between events (like the distance between successive potholes).
2. Type of distribution:
    - **Poisson:** A discrete distribution (since it deals with counts of events, which are whole numbers).
    - **Exponential:** A continuous distribution (since time or distance between events can take any positive value).
    
#### Relationship Between Them:   
- The Poisson distribution counts how many events happen in a given interval.
- The exponential distribution describes the time or distance until the next event happens.
- The exponential distribution is often the "waiting time" between events in a Poisson process.

#### In simple terms:
- If the number of events in a period follows a Poisson distribution,
- Then the time between events follows an exponential distribution.

     - a. Between 10 and 20 hours;
     - b. Less than two hours;
     - c. Longer than 50 hours.
- (ii)   State the mean and standard deviation of the intervals between successive signals.
- (iii)  Following a warning signal, what is the longest time the production process could be left unsupervised whilst ensuring the probability of missing the next signal is less than 0.01?

#### Solution
**Given:**<br>
The interval X between successive signals follows an exponential distribution with parameter 
X between successive signals follows an exponential distribution with parameter Î»=0.08.

(i) Probability calculations:<br>
We will use the cumulative distribution function (CDF) of the exponential distribution for this.
$$
\begin{aligned}
& \bullet \quad \text{CDF: } F(x) = 1 - e^{-\lambda x} \\[10pt]
& \bullet \quad \text{Survival function (1 - CDF): } P(X > x) = e^{-\lambda x}
\end{aligned}
$$

    - (a) Probability that the interval is between 10 and 20 hours:
    This is P(10<X<20)=F(20)âˆ’F(10).

```python
import numpy as np

# Given parameter lambda
lam = 0.08

# CDF of exponential distribution
def cdf(x, lam=0.08):
    return 1 - np.exp(-lam * x)

# Probability between 10 and 20 hours
prob_a = cdf(20, lam) - cdf(10, lam)

print(f'Probability that the interval is between 10 and 20 hours: {prob_a:,.2f}')
```

    - (b) Probability that the interval is less than 2 hours:
        This is P(X<2)=F(2).

```python
# Probability less than 2 hours
prob_b = cdf(2, lam)

print(f'Probability that the interval is less than 2 hours: {prob_b:,.2f}')
```

    - (c) Probability that the interval is longer than 50 hours:
        This is the survival function P(X>50)=1âˆ’F(50).

```python
# Probability longer than 50 hours
prob_c = 1 - cdf(50, lam)

print(f'Probability that the interval is longer than 50 hours: {prob_c:,.2f}')
```

(ii) Mean and standard deviation of the intervals:
For an exponential distribution:
$$
\begin{aligned}
& \bullet \quad \text{E(X) - Mean } \mu = \frac{1}{\lambda} \\[10pt]
& \bullet \quad \text{SD(X) - Standard deviation } ğœ = \frac{1}{\lambda}
\end{aligned}
$$
- For an exponential distribution, the mean Î¼ and standard deviation Ïƒ are both inversely related to the rate parameter and they are equal in value.
- This symmetry arises because the exponential distribution models the time between independent, random events that occur at a constant average rate, which leads to this balance in spread (standard deviation) and central tendency (mean).

```python
# Mean and standard deviation
mean = 1 / lam
std_dev = 1 / lam

mean, std_dev

print(f'Mean of the intervals: {mean:,.2f} hours') 
print(f'Standard deviation of the intervals: {std_dev:,.2f} hours') 
```

(iii) Longest time the process can be left unsupervised with less than 0.01 probability of missing the next signal:
We want to find the time t such that the probability of the interval being less than t (P(X<t) = 0.01). This means to calculate how long it takes for a signal to not occur with high probability  
$$
P(X<t) = 0.01)= 1 - e^{âˆ’0.08t} = 0.01 âŸ¹ e^{âˆ’0.08t} = 0.99
$$

```python
# how long it takes for a signal to not occur with high probability of P(X < t) = 0.01
t_hours = np.log(0.99) / -lam
t_minutes = t_hours * 60  # Convert to minutes
t_minutes
print(f'The production process should be unsupervised no longer than {t_minutes:,.2f} minutes to ensure the probability of missing a signal is less than 0.01')
```

```python
# The function calculates the means of multiple random samples from a population 
# and returns a list of these means. 
# This can be useful for estimating properties of a population through random sampling, 
# a common technique in statistical analysis.

def sample_mean_calculator(population_array, sample_size, n_samples):
    sample_means = [] # Initializes an empty list to store the means of the samples
    for i in range(n_samples): # A loop that runs n_samples times.
        sample = np.random.choice(population_array, size=sample_size, replace=False)
        # replace=False ensures that the elements are sampled without replacement, 
        # meaning once an element is chosen, it is not put back in the population for subsequent choices.
        sample_mean = stat.mean(sample) # calculates the mean of the sample
        sample_means.append(sample_mean) # The calculated mean is appended to the sample_means list.
    return sample_means
```

### Central Limit Theorem - Practice

200 fair dice are rolled and the results are added
- Calculate the mean and variance of a single random variable
- Using CLT, what is the probability that the sum is over 750? (Use norm.dist function in Excel)

```python
# Calculating the mean and variance of a single random variable
# Possible outcomes of a single die roll
outcomes = [1, 2, 3, 4, 5, 6]

# Calculate the mean of a single die roll
mean_single_die = sum(outcomes) / len(outcomes)

# Calculate the variance of a single die roll
expected_value_squared = sum([x**2 for x in outcomes]) / len(outcomes)
variance_single_die = expected_value_squared - mean_single_die**2

mean_single_die, variance_single_die
```

```python
## Using CLT, what is the probability that the sum is over 750?
import scipy.stats as stats

# Given values
n_dice = 200
target_sum = 750

# Mean and standard deviation of the sum of 200 dice rolls
mean_sum = n_dice * mean_single_die
variance_sum = n_dice * variance_single_die
std_dev_sum = variance_sum ** 0.5

# Calculate the Z-score for the sum of 750
z_score = (target_sum - mean_sum) / std_dev_sum

# Calculate the probability using the cumulative distribution function (CDF)
probability_over_750 = 1 - stats.norm.cdf(z_score)

probability_over_750
```

### Answer
The probability that the sum of 200 fair dice rolls is over 750 is approximately 1.93%.

### Central Limit Theorem - Practice
A company states that the time in minutes a customer has to wait until a call center representative picks up has an exponential distribution with a mean waiting time of 10 minutes (i.e. parameter Î» = 1/10 = 0.1

The company sampled 80 calls at random and found that the average waiting time was 12 minutes
- Using CLT, calculate the probability of waiting time being larger than 12 minutes

```python
# Given values
mean_population = 10  # population mean
# Standard deviation of the exponential distribution is given by the same as the mean:
std_dev_population = 10  # standard deviation for exponential distribution
sample_size = 80
sample_mean = 12

# Standard error
std_error = std_dev_population / (sample_size ** 0.5)

# Z-score
z_score_sample = (sample_mean - mean_population) / std_error

# Probability of observing a sample mean larger than 12 minutes
probability_over_12 = 1 - stats.norm.cdf(z_score_sample)

probability_over_12
```

- When performing hypothesis testing we have:
    - ğ»0 â€“ the â€œnull hypothesisâ€
    - ğ»1 â€“ the â€œalternative hypothesisâ€ (sometimes denoted as ğ»ğ´)
- The null hypothesis is the current state (click rates with the blue button) â€“ the existing belief we have about a parameter

- We need to prove that the alternative assumption is right:
    - If we can â€“ we reject ğ»0 (and therefore accept the alternative assumption)
    - If we cannot â€“ we do not reject ğ»0 (and therefore reject the alternative assumption)
    
- In order to decide which hypothesis we believe we perform a statistical test

- We ask ourselves how likely is it to see this result if we believe the null hypothesis?
- If it unlikely to see this results under the null hypothesis â€“ we get a good indication that it may not be reasonable to believe our original assumption â€“ so we can accept the alternative hypothesis

### P-value

P value is the probability of seeing what we saw in the sample under the null hypothesis

- P value is also the probability of rejecting the null hypothesis when it is true
    - We want this probability to be relatively small
- It is usually common to make decisions in comparison to a significance level of 5%
    - If P value of the test is smaller than 5% we reject ğ»0 (and accept the alternative hypothesis)
    - If P value is larger than 5% we stick with ğ»0 (and reject the alternative assumption)

## 8.4 <a id="examples_of_z_test">Examples of Z Test</a> <font size='3'>[[Table of contentsğŸ”]](#index)]<font size>

### One-Sample Z-Test   
    
 We perform two-samples Z-test when we we want to compare **a sample mean with the population mean**
    
 $$
\begin{gathered}
\overbrace{z \text{ score}}  = \frac{\overbrace{(\bar{x}_1 - \bar{x}_2)}^{\text{Difference bw}\\\text{Sample mean}\\\text{X}_1 - \text{X}_2} - \underbrace{(\mu_1 - \mu_2)}_{\text{Difference bw}\\\text{population mean}\\\mu_1 - \mu_2}}{\sqrt{\underbrace{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}}_{\text{Population}\\\text{standard}\\\text{deviation }\sigma_1, \sigma_2} \underbrace{\phantom{\frac{n_1}{n_2}}}_{\text{Sample Size}\\\text{n}_1, \text{n}_2}}}
\end{gathered}
$$   
    
    
Let's say we need to determine if girls on average score higher than 600 in the exam. We have the information that the standard deviation for girls' scores is 100. So, we collect the data of 20 girls by using random samples and record their marks. Finally, we also set our Î± value (significance level) to be 0.05.

**Assumptions**:

1. The null hypothesis would be that the mean score is 600.
2. The alternative hypothesis is that the mean score is higher than 600.
3. The known standard deviation is 100.
4. The sample size is 20.
5. The significance level (Î±) is 0.05.

This information would be used to conduct a one-tailed z-test, assuming the population is normally distributed or the sample size is large enough for the Central Limit Theorem to apply.

```python
# Load the CSV file
file_path = 'ZTest_Example.csv'
data_1 = pd.read_csv(file_path)

# Display the first few rows of the data to understand its structure
data_1.head()
```

```python
from scipy import stats
import numpy as np

# Known values
population_mean = 600
population_std_dev = 100
alpha = 0.05

# Sample data
sample = data_1['Score']
sample_size = len(sample)
sample_mean = np.mean(sample)

# Calculate z-score
z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# Calculate p-value (for one-tailed test)
p_value = 1 - stats.norm.cdf(z_score)

# Critical z-value for one-tailed test at alpha = 0.05
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha)

print(f"z_score: {z_score:,.4f}")
print(f"p_value: {p_value:,.4f}")
print(f"Critical Z-Value: {critical_value}")
print(f"Significance level: {alpha}")
print('='*50)

# Compare z-score with critical value
if z_score > critical_value:
    print("z score more then critical value, so we are rejecting the null hypothesis.")
else:
    print("z_score less or equal then critical_value, so we are failing to reject the null hypothesis.")
    
# Compare p-value with significance level (alpha)
if p_value < alpha:
    print("p-value less then significance level, so we are rejecting the null hypothesis.")
else:
    print("p-value more or equal then significance level, so we are failing to reject the null hypothesis.")
```

### Interpretation:
Since the p-value (0.0334) is less than the significance level (Î±=0.05), we reject the null hypothesis.<br>
Therefore, we have sufficient evidence to conclude that the average score of girls is significantly higher than 600.

### Two-Sample Z-Test   
    
Two-sample z-test formula (**for population standard deviations known**):
    
z = (xÌ„â‚ - xÌ„â‚‚) / âˆš(Ïƒâ‚Â²/nâ‚ + Ïƒâ‚‚Â²/nâ‚‚)

- xÌ„â‚, xÌ„â‚‚ are sample means

### What is the T-Test?
T-tests are a statistical way of testing a hypothesis when:

- We do not know the population variance
- Our sample size is small, n < 30    

### One-Sample T-Test

We perform a One-Sample t-test when we want to **compare a sample mean with the population mean**.<br> 
The difference from the z-Test is that **we do not have the information on Population Variance** here.<br> 
We use the **sample standard deviation** instead of population standard deviation in this case.

$$
\begin{gathered}
t = \frac{\overbrace{\bar{x} - \mu}^{\text{Sample mean } - \text{ Population mean}}}{\underbrace{\frac{s}{\sqrt{n}}}_{\text{Sample}\\\text{standard}\\\text{deviation}/\sqrt{\text{Sample Size}}}}
\end{gathered}
$$

Hereâ€™s an Example to Understand a One Sample T-Test

Letâ€™s say we want to determine if on average girls score more than 600 in the exam. We do not have the information related to variance (or standard deviation) for girlsâ€™ scores. To a perform t-test, we randomly collect the data of 10 girls with their marks and choose our âº value (significance level) to be 0.05 for Hypothesis Testing.

In this example:

- Mean Score for Girls is 606.8
- The size of the sample is 10
- The population mean is 600
- Standard Deviation for the sample is 13.14

**Assumptions:**
1. Null hypothesis (H0): The average score of girls is less than or equal to 600.
2. Alternative hypothesis (H1): The average score of girls is greater than 600.
3. Significance level (Î±): 0.05.

```python
# Load the CSV file
file_path = 'T_Test_Example.csv'
data_t_1 = pd.read_csv(file_path)

# Display the first few rows of the data to understand its structure
data_t_1.head()
```

```python
# Sample data
girls_scores = data_t_1['Girls_Score']

# Hypothesized population mean
population_mean = 600

# Sample size
n = len(girls_scores)

# Sample mean
sample_mean = np.mean(girls_scores)

# Sample standard deviation
sample_std_dev = np.std(girls_scores, ddof=1)  # ddof=1 provides sample standard deviation

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(girls_scores, population_mean)

# Adjust for one-tailed test
if t_statistic > 0:
    p_value /= 2

# Critical value for one-tailed t-test at alpha = 0.05
alpha = 0.05
df = n - 1  # degrees of freedom

critical_value = stats.t.ppf(1 - alpha, df)

# Display results
print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")
print(f"Critical T-Value: {critical_value}")
print(f"Significance level: {alpha}")
print('='*50)

# Compare p-value with significance level (alpha)
if p_value < alpha:
    print("p-value less then significance level, so we are rejecting the null hypothesis. The average score is significantly greater than 600.")
else:
    print("p-value more or equal then significance level, so we are failing to reject the null hypothesis. There is no significant evidence that the average score is greater than 600.")

# Compare t-score with critical value
if t_statistic > critical_value:
    print("t_statistic more than critical value, so we are rejecting the null hypothesis. The average score is significantly greater than 600.")
else:
    print("t_statistic less or equal than critical value, so we are failing to reject the null hypothesis. There is no significant evidence that the average score is greater than 600.")
```

### Two-Sample T-Test

Two-sample t-test formula **(for population standard deviations unknown)**:

t = (xÌ„â‚ - xÌ„â‚‚) / âˆš(sâ‚Â²/nâ‚ + sâ‚‚Â²/nâ‚‚)

where:

- xÌ„â‚, xÌ„â‚‚ are sample means
- Ïƒâ‚, Ïƒâ‚‚ are known population standard deviations
- nâ‚, nâ‚‚ are sample sizes

Hereâ€™s an Example to Understand a Two-Sample T-Test

Here, letâ€™s say we want to determine if on average, boys score 15 marks more than girls in the exam. We do not have the information related to variance (or standard deviation) for girlsâ€™ scores or boysâ€™ scores. To perform a t-test. we randomly collect the data of 10 girls and boys with their marks. We choose our âº value (significance level) to be 0.05 as the criteria for Hypothesis Testing.

In this example:

- Mean Score for Boys is 630.1
- Mean Score for Girls is 606.8
- Difference between Population Mean 15

**Assumptions:**
1. Null Hypothesis (H0): 
    - H0: Î¼_boys - Î¼_girls â‰¤ 15
    - In words: "Boys do not score more than 15 marks higher than girls"
2. Alternative Hypothesis (H1 ): 
    - H1: Î¼_boys - Î¼_girls > 15
    - In words: "Boys score more than 15 marks higher than girls"
    - This is a one-tailed test (right-tailed)
3. Significance level (Î±): 0.05.    

```python
# Load the CSV file
file_path = 'T_Test_Example_2_samples.csv'
data_t_2 = pd.read_csv(file_path)

# Display the first few rows of the data to understand its structure
data_t_2.head()
```

```python
import numpy as np
from scipy import stats

# Data
girls_scores = [587, 602, 627, 610, 619, 622, 605, 608, 596, 592]
boys_scores = [626, 643, 647, 634, 630, 649, 623, 625, 617, 607]

# Step 1: Calculate basic statistics
n1 = len(girls_scores)
n2 = len(boys_scores)
mean_girls = np.mean(girls_scores)
mean_boys = np.mean(boys_scores)
std_girls = np.std(girls_scores, ddof=1)
std_boys = np.std(boys_scores, ddof=1)

# Step 2: Set up hypotheses and test value
# H0: Î¼_boys - Î¼_girls â‰¤ 15
# H1: Î¼_boys - Î¼_girls > 15
test_value = 15
alpha = 0.05

# Step 3: Calculate t-statistic
pooled_std = np.sqrt((std_girls**2/n1) + (std_boys**2/n2))
t_stat = ((mean_boys - mean_girls) - test_value) / pooled_std

# Step 4: Calculate degrees of freedom (Welch-Satterthwaite equation)
df = ((std_girls**2/n1 + std_boys**2/n2)**2) / \
     ((std_girls**4/(n1**2*(n1-1)) + std_boys**4/(n2**2*(n2-1))))

# Step 5: Find critical value (one-tailed)
t_crit = stats.t.ppf(1 - alpha, df)  # right-tailed test

# Step 6: Calculate p-value (one-tailed)
p_value = 1 - stats.t.cdf(t_stat, df)  # right-tailed test

# Print results
print("Test Statistics:")
print(f"Mean difference (boys - girls): {mean_boys - mean_girls:.2f}")
print(f"Sample standard deviation (girls): {std_girls:.2f}")
print(f"Sample standard deviation (boys): {std_boys:.2f}")
print(f"Degrees of freedom: {df:.2f}")
print(f"\nT-statistic: {t_stat:.4f}")
print(f"Critical value: {t_crit:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Significance level: {alpha}")

# Decision using critical value
print("\nMethod 1 - Using Critical Value:")
if t_stat > t_crit:
    print("Reject H0: There is significant evidence that boys score more than 15 marks higher than girls")
else:
    print("Fail to reject H0: There is not enough evidence that boys score more than 15 marks higher than girls")

# Decision using p-value
print("\nMethod 2 - Using P-value:")
if p_value < alpha:
    print("Reject H0: There is significant evidence that boys score more than 15 marks higher than girls")
else:
    print("Fail to reject H0: There is not enough evidence that boys score more than 15 marks higher than girls")
```

### Interpretation
At a 5% significance level, we do not have enough evidence to conclude that boys score more than 15 marks higher than girls, even though the sample mean difference is 24.30 marks. The p-value (0.0642) is close to our significance level (0.05) in this one-tailed test, but still not small enough to reject H0.

#### Chi-Square Statistic Formula

The chi-square statistic measures the difference between observed and expected counts:

### Question 3
In examining the invoices issued by a company, an auditor finds that the dollar amount of invoices have a mean of \\$1,732 and standard deviation of $298. Which pair of symmetric numbers around the mean make the statement P(a<XÌ„<b)=0.853 correct for a random sample of 55 invoices?

```python
import scipy.stats as stats
import numpy as np

# Given parameters
population_mean = 1732
population_std = 298
n = 55  # sample size
desired_probability = 0.853

# For sample mean:
# Standard error of the mean
standard_error = population_std / np.sqrt(n)

# Since we want symmetric bounds around the mean, and the total probability is 0.853,
# we need (1 - 0.853)/2 = 0.0735 in each tail

# Find the z-score for the desired probability (one side)
# We want the z-score that gives us (1 + 0.853)/2 = 0.9265 probability
z_score = stats.norm.ppf((1 + desired_probability)/2)

# Calculate the bounds
margin = z_score * standard_error
lower_bound = population_mean - margin
upper_bound = population_mean + margin

print(f"Standard Error: ${standard_error:.2f}")
print(f"Z-score for {desired_probability:.3f} probability: {z_score:.4f}")
print(f"\nSymmetric bounds around mean:")
print(f"Lower bound (a): ${lower_bound:.2f}")
print(f"Upper bound (b): ${upper_bound:.2f}")

# Verify our answer
prob = stats.norm.cdf(upper_bound, population_mean, standard_error) - \
       stats.norm.cdf(lower_bound, population_mean, standard_error)
print(f"\nVerification - Probability between bounds: {prob:.4f}")
```

### Interpretation:
The interpretation is that for 85.3% of all possible samples of 55 invoices, the sample mean will fall between these two bounds.
